```python
# 0. Dependent & Independent Parameters

import pandas as pd
data = {
    'independent_param1': [1,2,3,4,5],
    'independent_param2': [10,20,30,40,50],
    'dependent_param': [100,200,300,400,500],
}

df = pd.DataFrame(data)

independent_params = df.drop('dependent_param', axis=1)
dependent_param = df['dependent_param']

print('Independent Parameters: ')
print(independent_params)

print('Dependent Parameters: ')
print(dependent_param)
```

```python
# 1. Data Visualization Libraries

# 1. Matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()

# 2. Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.show()

# 3. Plotly
import plotly.express as px
iris = px.data.iris()
fig = px.scatter(iris, x='petal_length', y='petal_width', color='species')
fig.show()
```

```python
# Regex
import re

pattern = r'\d+-\d+-\d+'
text = 'My date of birth is 07-01-1999 and her date of birth is 11-11-2002'
matches = re.findall(pattern, text)
print(matches)
```

```python
# Visualize Network Data

import networkx as nx
import matplotlib.pyplot as plt

g = nx.random_geometric_graph(11,0.4)

plt.figure(figsize=(8,6))
nx.draw(g, with_labels=True, node_size=300, node_color='skyblue', font_size=10, font_weight='bold')
plt.title('Network Data Visualization')
plt.show()
```

```python
# 1. Exploring Data Visualization

x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.figure(figsize=(8,6))
plt.plot(x,y, marker='o', linestyle='-', color='b', label='Prime Numbers')
plt.title('Data Visualization Framework')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 4. Scatter Plots

import folium
import pandas as pd

data = pd.DataFrame({
    'Latitude': [40.7128, 34.0522, 41.8781, 29.7604],
    'Longitude': [-74.0060, -118.2437, -87.6298, -95.3698],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Population': [8398748, 3990456, 2705994, 2320268]
})

map = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

for index, row in data.iterrows():
    folium.CircleMarker(location=(row['Latitude'], row['Longitude']),
                        radius=row['Population']/100000,
                        color='blue',
                        fill=True,
                        fill_color='blue',
                        fill_opacity=0.4,
                        tooltip=row['City'] + ': '+str(row['Population'])).add_to(map)
                        
map.save('scatter_plot_map.html')
                       
                       
                       
```
